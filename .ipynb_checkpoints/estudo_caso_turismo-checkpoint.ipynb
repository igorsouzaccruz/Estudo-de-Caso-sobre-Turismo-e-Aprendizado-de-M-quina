```python
chegadas_por_mes = turistas_internacionais_no_ceara.groupby('Mês')['Chegadas'].sum()
chegadas_por_mes
```

```python
# 1. Converter a coluna 'Mês' para datetime, se ainda não for
turistas_internacionais_no_ceara['Mês'] = pd.to_datetime(turistas_internacionais_no_ceara['Mês'])

# 2. Agrupar por mês e somar
chegadas_por_mes = turistas_internacionais_no_ceara.groupby('Mês')['Chegadas'].sum()

# 3. Garantir que o índice seja uma série temporal
chegadas_por_mes = chegadas_por_mes.sort_index()  # ordena cronologicamente
```

```python
chegadas_por_mes.plot(figsize=(12, 5), title="Chegadas de Turistas Internacionais por Mês")
plt.xlabel("Data")
plt.ylabel("Número de Chegadas")
plt.grid(True)
plt.show()
```

```python
# Decomposição da série temporal, criando uma variável para cada formato
chegadas_por_mes = chegadas_por_mes.asfreq('MS')  # 'MS' = mês no início do mês

decomposicao = seasonal_decompose(chegadas_por_mes, model='additive')  # ou model='multiplicative'
decomposicao.plot()
plt.show()
```
